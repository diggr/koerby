```python
import json
import pandas as pd
```

```python
with open("import/giantbomb_companies.json") as f:
    data = json.load(f)
```

```python
data[0].keys()
```

```python
export = []
for company in data:
    if company["location_country"] or company["website"]:
        export.append(company)
```

```python
len(export)
```

```python
df = pd.DataFrame(export)
```

```python
df
```

```python
df[["deck","guid", "location_address", "location_city", "location_country", "location_state", "name", "phone", "website", "aliases" ]].to_csv("source/giantbomb_companies.csv", index=False)
```
